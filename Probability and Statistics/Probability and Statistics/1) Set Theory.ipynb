# Set Theory

* Empty set - $\phi$

$2 \in \{1,2,3\}$

* This means that the element 2 exists in the set of 1, 2, and 3


$4 \not\in {1,2,3}$
* This means that the element 4 does not exist in the set of 1, 2, and 3


$A = \{1,5,6\}, B = \{1,5,6,7,8\}$

A is a subset of B and B is called the superset of A.

$A \subseteq B$

```python
def f_issubset(A,B):
    for e in A:
        if e in B:
            pass
        else:
            return False
    return True
```

```python
print(f_issubset({1,5,6},{1,2, 4,5,6, 7, 8, 9}))
```

The union of two sets forms a new set containing those elements in sets A or B.

A $\cup$ B

```python
import numpy as np
A = set(np.arange(0,9,2))
A
```

```python
B = set(np.arange(1,9,3))
B
```

```python
A.union(B)
```

The intersection of two sets froms a new set containing only those elements belonging to both A and B. 

A $\cap$ B

```python
A.intersection(B)
```

The difference between sets A and B is the set of all elements in A, but not B, denoted $ A - B = x \in A | x \not \in B$

```python
A.difference(B)
```

```python
U = set(np.arange(10))
U
```

$ U - (A \cup B)$

```python
U.difference(A.union(B))
```

$ (U - A) \cap (U - B)$

```python
U.difference(A).intersection(U.difference(B))
```

$ U - (A \cap B)$

```python
U.difference(A.intersection(B))
```

$ (U - A) \cup (U - B)$ 

```python
U.difference(A).union(U.difference(B))
```

## Jaccard Similarity

$JS(A,B) = \frac{A\cap B}{A\cup B}$

```python
len(A.intersection(B)) / len(A.union(B))
```
